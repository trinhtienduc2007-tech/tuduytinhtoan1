```python
Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
```

```python
h = int(input("Nhập h: "))
for i in range(1, h + 1):
    for j in range(1, 2 * h):
        if i == h or j == h - i + 1 or j == h + i - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên

```python
a = int(input("Nhập a: "))
check = False
for i in range(1, a + 1):
    if i * i == a:
        check = True
        break
print("Là số chính phương" if check else "Không phải số chính phương")
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó

```python
a = int(input("Nhập a: "))
tong_uoc = 0
for i in range(1, a):
    if a % i == 0:
        tong_uoc += i
print("Là số hoàn hảo" if tong_uoc == a else "Không phải số hoàn hảo")
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n 


```python
n = int(input("Nhập n: "))
s = 0
k = 0
while s + (k + 1) < n:
    k += 1
    s += k
print("Số k tìm được là:", k)
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A


```python
A = float(input("Nhập A: "))
s = 0
n = 0
while s <= A:
    n += 1
    s += 1/n
print("Số n nhỏ nhất là:", n)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
max_val = -float('inf')
min_val = float('inf')
while True:
    num = int(input("Nhập số (-1 để dừng): "))
    if num == -1: break
    if num > max_val: max_val = num
    if num < min_val: min_val = num

if max_val == -float('inf'):
    print("Chưa nhập số nào hợp lệ.")
else:
    print(f"Max: {max_val}, Min: {min_val}")
```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng. 

```python
n = int(input("Nhập n: "))
temp = abs(n)
count = 0
if temp == 0: count = 1
while temp > 0:
    count += 1
    temp //= 10
print("Số chữ số:", count)
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 


```python
n = abs(int(input("Nhập n: ")))
chan = 0
le = 0
if n == 0: chan = 1
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0: chan += 1
    else: le += 1
    n //= 10
print(f"Chẵn: {chan}, Lẻ: {le}")
```

Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n 


```python
n = abs(int(input("Nhập n: ")))
tong = 0
while n > 0:
    tong += n % 10
    n //= 10
print("Tổng các chữ số:", tong)
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không 

```python
n = int(input("Nhập n: "))
temp = n
is_3k = True
if n < 1: 
    is_3k = False
else:
    while temp > 1:
        if temp % 3 != 0:
            is_3k = False
            break
        temp //= 3
print(f"Có dạng 3^k" if is_3k else "Không có dạng 3^k")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1
result = 1
while f2 <= A:
    result = f2
    f1, f2 = f2, f1 + f2
print("Số Fibonacci lớn nhất không quá A là:", result)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = abs(int(input("Nhập n: ")))
count = 0
while n > 0:
    d = n % 10
    if d == 2 or d == 3 or d == 5 or d == 7:
        count += 1
    n //= 10
print("Số lượng chữ số là số nguyên tố:", count)
```
