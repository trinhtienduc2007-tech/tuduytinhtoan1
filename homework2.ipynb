Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó.

```python
a = float(input("Nhập chiều dài: "))
b = float(input("Nhập chiều rộng: "))
chu_vi = (a + b) * 2
dien_tich = a * b
print(f"Chu vi: {chu_vi}, Diện tích: {dien_tich}")
```

Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.

```python
import math
r = float(input("Nhập bán kính hình tròn: "))
chu_vi = 2 * math.pi * r
dien_tich = math.pi * r**2
print(f"Chu vi: {chu_vi:.2f}, Diện tích: {dien_tich:.2f}")
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
import math

a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        loai = "Tam giác đều"
    elif a == b or b == c or a == c:
        if (a**2 + b**2 == round(c**2, 2)) or (a**2 + c**2 == round(b**2, 2)) or (b**2 + c**2 == round(a**2, 2)):
            loai = "Tam giác vuông cân"
        else:
            loai = "Tam giác cân"
    elif (a**2 + b**2 == round(c**2, 2)) or (a**2 + c**2 == round(b**2, 2)) or (b**2 + c**2 == round(a**2, 2)):
        loai = "Tam giác vuông"
    else:
        loai = "Tam giác thường"
    
    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    print(f"Đây là {loai}. Chu vi: {chu_vi}, Diện tích: {dien_tich:.2f}")
else:
    print("Khong la ba canh cua mot tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a == 0:
    # Trở thành pt bậc 1
    if b == 0:
        print("Vô nghiệm" if c != 0 else "Vô số nghiệm")
    else:
        print(f"x = {-c/b}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        print(f"Nghiệm kép x = {-b / (2*a)}")
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f"x1 = {x1}, x2 = {x2}")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math
x0 = float(input("Tọa độ x tâm đường tròn: "))
y0 = float(input("Tọa độ y tâm đường tròn: "))
r = float(input("Bán kính R: "))
xa = float(input("Tọa độ x điểm A: "))
ya = float(input("Tọa độ y điểm A: "))

khoang_cach = math.sqrt((xa - x0)**2 + (ya - y0)**2)

if math.isclose(khoang_cach, r):
    print("Điểm A thuộc đường tròn")
elif khoang_cach < r:
    print("Điểm A nằm trong đường tròn")
else:
    print("Điểm A nằm ngoài đường tròn")
```
